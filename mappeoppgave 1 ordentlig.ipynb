```python
def fig2(): 
    Q_values = np.linspace(1, 4000, 4000)
    MR_values = marginal_revenue(Q_values)
    MC_values = marginal_cost(Q_values)
    demand_values = demand(Q_values)
    
    plt.plot(Q_values, MR_values, label='MR')
    plt.plot(Q_values, MC_values, label='MC')
    plt.plot(Q_values, demand_values, color='green', label='Ettersp√∏rsel')
    
    Q_star = np.argmin(np.abs(MR_values - MC_values))
    P_star = demand(Q_star)

    Q_opt = np.argmin(np.abs(demand_values - MC_values))
    P_opt = demand(Q_opt)
    
    KO = 0.5 * (demand(0) - P_star) * Q_star
    PO = 0.5 * (P_star - MC_values[Q_star]) * Q_star
    D√òD = 0.5 * (P_opt - P_star) * (Q_opt - Q_star)
    
    plt.fill_between(Q_values[:Q_star+1], P_star, demand_values[:Q_star+1], color='blue', alpha=0.3, label='KO')
    plt.fill_between(Q_values[:Q_star+1], MC_values[:Q_star+1], P_star, color='red', alpha=0.3, label='PO')
    plt.fill_between(Q_values[Q_star:Q_opt+1], MC_values[Q_star:Q_opt+1], demand_values[Q_star:Q_opt+1], color='yellow', alpha=0.3, label='D√∏dvektstap')

    plt.plot(Q_equ, p_eq, marker='o', color='red', label='Likevekt')
    plt.plot([Q_equ, Q_equ], [0, p_eq], linestyle='--', color='black')
    plt.plot([0, Q_equ], [p_eq, p_eq], linestyle='--', color='black')
    
    plt.legend(loc='upper right')
    plt.title("Hurtigruten monopol: KO, PO og D√∏dvektstap")
    plt.ylabel('Pris')
    plt.xlabel('Lugarer')
    plt.ylim((0, 10000))
    plt.xlim((0, 4000)) 
    plt.show()
    
    print(f"Konsumentoverskudd: {KO:.2f}")
    print(f"Produsentoverskudd: {PO:.2f}")
    print(f"D√∏dvektstap tap: {D√òD:.2f}")

fig2()
```

### Stackelberg modell

```python
q1, q2,c, a, b, pi,i=symbols('q1 q2 c a b pi i')

def P_demand(Q,a,b):
    return a-b*Q

def profit(q1,q2,c,a,b):
    return (P_demand(q1+q2,a,b)-c)*q1
```

Vi dereiverer profittfunksjon til bedrift 2 mhp q2: œÄ2 = (P-c)*q2=(a-b(q1+q2)-c)*q2

```python
d_profit2_Q=diff(profit(q2,q1,c,a,b),q2)
d_profit2_Q
```

Setter den derivert lik 0 og finner reaksjonsfunksjon til bedrift 2

```python
Q2_sol1=solve(d_profit2_Q,q2)[0]
Q2_sol1
```

P√• trinn 1 sette vi reaksjonsfunksjonene til bedrift 2 inn i bedrift 1 sin profittfunksjon, og deriverer dette utrykket mhp q1. Maks œÄ1 =(a-b(q1+q2)-c)*q1 gitt q2=(a-bq1-c)/2b

```python
d_profit1_Q=diff(profit(q1,Q2_sol1,c,a,b),q1)
d_profit1_Q
```

For √• finne optimalt kvantum til lederbedriften setter vi uttrykket over lik 0

```python
Q1_sol=solve(d_profit1_Q,q1)[0]
Q1_sol
```

Vi setter s√• optimalt valg av q1 inn i reaksjonsfunksjonen til bedrift 2

```python
Q2_sol2=Q2_sol1.subs({q1:Q1_sol})
Q2_sol2
```

```python
stackelberg=lambdify(
    (a,b,c), 
    (Q1_sol,Q2_sol2))
```

#### Pris monopol og Stackelberg

```python
def monopol_price(Am, cm):
    Ps = (Am + cm)/ 2
    return Ps

Am = 9360
cm = 1000

price = monopol_price(Am, cm)
print("monopol pris:", price)
```

```python
def stackelberg_price(As, cs):
    Ps = (As + 3 * cs) / 4
    return Ps

As = 9360
cs = 1000

price = stackelberg_price(As, cs)
print("Stackelberg price:", price)
```

```python

q1, q2, c, a, b, pi, i = sp.symbols('q1 q2 c a b pi i')

def P_demand(Q, a, b):
    return a - b * Q

def profit(q1, q2, c, a, b):
    return (P_demand(q1 + q2, a, b) - c) * q1

d_profit2_Q = sp.diff(profit(q2, q1, c, a, b), q2)
Q2_sol1 = sp.solve(d_profit2_Q, q2)[0]

d_profit1_Q = sp.diff(profit(q1, Q2_sol1, c, a, b), q1)
Q1_sol = sp.solve(d_profit1_Q, q1)[0]

Q2_sol2 = Q2_sol1.subs({q1: Q1_sol})

stackelberg = sp.lambdify((a, b, c), (Q1_sol, Q2_sol2))

Q1_opt, Q2_opt = stackelberg(9360, 2, 1000)
Q_optimal = Q1_opt + Q2_opt

optimal_price = P_demand(Q_optimal, 9360, 2)


Q_values = np.linspace(0, 5000, 1000)
P_values = P_demand(Q_values, 9360, 2)


def fig_test(): 
    Q_values = np.linspace(0, 5000, 1000)
    P_values = P_demand(Q_values, 9360, 2)
    MC_values = marginal_cost(Q_values)

    plt.plot(Q_values, P_values, label='Ettersp√∏rsel')
    plt.plot(Q_values, MC_values, color='orange', label='MC', linestyle='solid')
    
    plt.plot([0, Q_optimal], [optimal_price, optimal_price], 'g--', label='Optimal pris')
    plt.plot([Q_optimal, Q_optimal], [0, optimal_price], 'r--', label='Optimalt kvantum')
    
    plt.plot(Q_optimal, optimal_price, marker='o', color='black', label='Stackelberg likevekt')
    
    plt.xlabel('Kvantum')
    plt.ylabel('Pris')
    plt.title('Likevekt ved optimalt kvantum og optimal pris ved Stackelberg')
    plt.ylim((0, 10000))
    plt.xlim((0, 4000)) 
    plt.legend()
    plt.show()

fig_test()

```

```python
def producer_surplus(Q, a, b, MC):
    return (Q * (P_demand(Q, a, b) - MC))

PS_optimal = producer_surplus(Q_optimal, 9360, 2, marginal_cost(Q_optimal))

def fig_test():
    Q_values = np.linspace(0, 5000, 1000)
    P_values = P_demand(Q_values, 9360, 2)
    MC_values = marginal_cost(Q_values)

    plt.plot(Q_values, P_values, label='Ettersp√∏rsel')
    plt.plot(Q_values, MC_values, color='orange', label='MC', linestyle='solid')

    plt.plot([0, Q_optimal], [optimal_price, optimal_price], 'g--', label='Optimal pris')
    plt.plot([Q_optimal, Q_optimal], [0, optimal_price], 'r--', label='Optimalt kvantum')

    plt.plot(Q_optimal, optimal_price, marker='o', color='black', label='Stackelberg likevekt')

    plt.fill_between(Q_values, P_values, optimal_price, where=(Q_values <= Q_optimal), color='blue', alpha=0.3, label='Konsumentoverskudd')

    plt.fill_between(Q_values, MC_values, optimal_price, where=(Q_values <= Q_optimal), color='red', alpha=0.3, label='Produsentoverskudd')

    plt.fill_between(Q_values, P_values, MC_values, where=(Q_values > Q_optimal), color='yellow', alpha=0.3, label='D√∏dvektstap')

    plt.xlabel('Kvantum')
    plt.ylabel('Pris')
    plt.title('Likevekt ved optimalt kvantum og optimal pris ved Stackelberg')
    plt.ylim((0, 10000))
    plt.xlim((0, 4000))
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()

fig_test()
```
